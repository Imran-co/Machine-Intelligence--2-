5) Use a dataset (e.g., Iris) and implement a multi-class linear classifier using One-Vs-
Rest strategy. Evaluate accuracy.

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris

iris_data = load_iris()

X, y = iris_data.data, iris_data.target

```

```python
df = pd.DataFrame(X, columns=iris_data.feature_names)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LinearRegression


one_rest = OneVsRestClassifier(LinearRegression())

one_rest.fit(X_train, y_train)
```

```python
y_predict = one_rest.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
print("Accuracy of the model : ",accuracy_score(y_test, y_predict))
```
