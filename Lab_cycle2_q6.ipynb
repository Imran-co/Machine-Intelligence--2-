6) For a 1D dataset, estimate the mean and variance of the Gaussian PDF. Plot the data
and the estimated curve.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
np.random.seed(42)  # For reproducibility
true_mean = 5.0
true_std = 2.0
df = np.random.normal(loc=true_mean, scale=true_std, size=1000)
```

```python
df
```

```python
sample_mean = np.mean(df)
sample_std = np.std(df)
sample_var = np.var(df)
```

```python
print(f"True parameters: mean = {true_mean:.2f}, std = {true_std:.2f}")
print(f"Estimated parameters: mean = {sample_mean:.2f}, std = {sample_std:.2f}")
```

```python
print(f"Estimated variance = {sample_var:.2f}")
```

```python
plt.figure(figsize=(10, 6))
# plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Data histogram')

# Plot the PDF of the fitted normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, sample_mean, sample_std)
plt.plot(x, p, 'k', linewidth=2, label='Fitted PDF')

plt.title('Histogram of Data and Fitted Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency/Density')
plt.legend()
plt.show()
```

```python
x = np.linspace(min(df), max(df), 1000)
true_pdf = norm.pdf(x, true_mean, true_std)
plt.plot(x, true_pdf, 'r-', lw=2, alpha=0.6, label='True PDF')

# Plot estimated PDF
estimated_pdf = norm.pdf(x, sample_mean, sample_std)
plt.plot(x, estimated_pdf, 'b-', lw=2, label='Estimated PDF')

# Add labels and legend
plt.title('Gaussian PDF Estimation from 1D Data')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
